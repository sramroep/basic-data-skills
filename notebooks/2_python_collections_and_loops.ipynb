## 2.3 Safe access with `.get()`

If you access a key that doesn't exist, Python raises a `KeyError`.
We'll first see that error, then use `.get()` for safer access.


```python
country = {"name": "Netherlands", "capital": "Amsterdam"}

# This intentionally shows a KeyError without stopping the notebook.
try:
    print(country["currency"])
except KeyError as error:
    print(f"KeyError: {error}")
```

When you use square brackets (`[]`), the key must exist.
With `.get()`, missing keys return `None` (or a fallback value you choose).


```python
country = {"name": "Netherlands", "capital": "Amsterdam"}

print(country.get("currency"))              # None (no crash)
print(country.get("currency", "Unknown"))   # "Unknown" (fallback value)
print(country.get("name", "Unknown"))       # "Netherlands"
```

## 2.4 Nesting: real data is layered

Real-world data is usually nested — dictionaries inside dictionaries, lists of dictionaries, and so on.

**Tip:** Read nested access from left to right. `country["name"]["common"]` means: start with `country`, go into `"name"`, then go into `"common"`.

```python
country = {
    "name": {
        "common": "Netherlands",
        "official": "Kingdom of the Netherlands"
    },
    "capital": ["Amsterdam"],
    "languages": {
        "nld": "Dutch"
    },
    "borders": ["BEL", "DEU"]
}

# Navigate step by step
print(country["name"])                 # {"common": "Netherlands", ...}
print(country["name"]["common"])       # "Netherlands"
print(country["capital"][0])           # "Amsterdam"
print(country["borders"][1])           # "DEU"
```

## 2.5 This is basically JSON

`JSON` stands for **JavaScript Object Notation**.
It's a simple text format used to exchange structured data between systems, especially in APIs.

When Python reads JSON, it turns it into the same dictionaries and lists you've already practiced.


```python
# This is what an API sends over the internet (a text string):
json_text = '{"name": "Netherlands", "capital": ["Amsterdam"]}'
print(type(json_text))  # <class 'str'> — just text!

# Python converts it to a dictionary you can work with:
import json
data = json.loads(json_text)
print(type(data))       # <class 'dict'> — a dictionary!
print(data["name"])     # "Netherlands"
```

The conversion happens automatically when you call `response.json()` later. The key differences in the text format:

| Python | JSON text |
|--------|----------|
| `True` / `False` | `true` / `false` |
| `None` | `null` |
| Single or double quotes | Double quotes only |

You rarely need to worry about these differences — Python handles the conversion.

## 2.6 Reading and writing JSON files

Sometimes you'll want to save or load JSON files directly.

After running this cell, Python creates `example.json` in your notebook's current working folder.
Use the printed **Full path** to see the exact location.
- In the notebook UI: open the file browser and refresh.
- In your IDE: use the project file tree or search for `example.json`.


```python
import json
from pathlib import Path

# Save data to a JSON file
data = {"name": "Netherlands", "capital": "Amsterdam", "population": 17000000}

with open("example.json", "w") as f:
    json.dump(data, f, indent=2)

print("Saved to example.json")
print(f"Full path: {Path('example.json').resolve()}")

# Read it back
with open("example.json") as f:
    loaded = json.load(f)

print(f"Loaded: {loaded['name']}")
```

## 2.7 Loops: doing something for each item

A `for` loop repeats the same action for each item in a collection.

- Looping through a **list** gives you each item.
- Looping through a **dictionary** gives you each key by default.
- Use `.items()` when you want both key and value.


```python
countries = [
    {"name": "Netherlands", "capital": "Amsterdam", "population": 17000000},
    {"name": "Belgium", "capital": "Brussels", "population": 11000000},
    {"name": "Germany", "capital": "Berlin", "population": 83000000}
]

print("Looping through a list of dictionaries:")
for country in countries:
    print(country["name"])

country_pop_millions = {
    "Netherlands": 17.0,
    "Belgium": 11.0,
    "Germany": 83.0,
}

print("\nLooping through dictionary keys:")
for country_name in country_pop_millions:
    print(country_name)

print("\nLooping through dictionary key-value pairs:")
for country_name, pop_m in country_pop_millions.items():
    print(f"{country_name}: {pop_m} million")
```

## 2.8 Building new lists with loops

A very common pattern: loop through data and collect results into a new list using `.append()`.

```python
countries = [
    {"name": "Netherlands", "population": 17000000},
    {"name": "Belgium", "population": 11000000},
    {"name": "Germany", "population": 83000000}
]

# Extract just the names into a new list
names = []
for country in countries:
    names.append(country["name"])

print(names)  # ["Netherlands", "Belgium", "Germany"]
```

```python
# Build a simplified version of the data
summaries = []
for country in countries:
    summaries.append({
        "name": country["name"],
        "pop_millions": country["population"] / 1000000
    })

print(summaries)
```
